```scala
import org.apache.spark.sql.SparkSession
import java.util.Properties
import org.apache.spark.sql.types.{IntegerType, StringType, StructField, StructType}
import org.apache.spark.sql.{Row, SQLContext}
import org.apache.spark.{SparkConf, SparkContext}
```

## 1. Creating a DataFrame 

```scala
// Create a DataFrame from JSON file
val spark = SparkSession.builder().appName("Spark SQL basic example").config("spark.some.config.option", "read JSON").getOrCreate()
// need to import implicits._ which implicitly convert data into dataframe
import spark.implicits._
//define a dataframe dfSales
val dfSales = spark.read.option("multiline", true).json("sales_json.json")
println(dfSales.getClass)

dfSales.printSchema()

dfSales.show(5)
```

## 2. Operations on DataFrame

```scala
// Display a specific column, which is equivilent to SELECT [col] FROM TABLE
dfSales.select("company").show()
```

```scala
// Display more than one columns and switch the columen order
dfSales.select("country", "company").show()
```

```scala
// display rows with a filter:
//dfSales.filter(dfSales("id")>20).show(5)

// use full name reference is not very convenient, so you can use $ instead
dfSales.filter($"id">20).show(2)
// $ sign can be used to quote a field or a variable
// derived column.
dfSales.select($"id"+5).show()
```

```scala
// Use groupBy() method to show aggregated results
dfSales.groupBy("country").count().show()
```

```scala
// sorting
// sort by a column (asceding by default)
//dfSales.sort("country").show()


// sort by a column in a descending order
//dfSales.sort(desc("country")).show()
//dfSales.sort(dfSales("id").desc).show()
//dfSales.sort($"id".desc).show()
//dfSales.sort($"country".asc,$"id".desc).show()

// secondary sorting
dfSales.sort($"country".desc, $"id".asc).show()

// display the top 20 countries in sales (ascending and descending)
dfSales.groupBy("country").count().sort("count").show()
dfSales.groupBy("country").count().sort($"count".desc).show()
```

```scala
// The sql function on a SparkSession enables applications to run SQL queries programmatically and returns the result as a DataFrame.
dfSales.createOrReplaceTempView("saleTable")

val sqlDF = spark.sql("SELECT * FROM saleTable WHERE country = \"Japan\" ")
sqlDF.show()
// tempview is only valid within the session.
//spark.newSession().sql("SELECT * FROM saleTable WHERE id<=5").show()
```

## 3. Prgrammitical Interface 

```scala
import org.apache.spark.sql.types._

val peopleRDD = sc.textFile("people.txt").map(x=>x.split(','))
// Step 1: Convert records of the RDD (people) to Rows
val rowRDD =  peopleRDD.map(x => Row(x(0), x(1).trim))

// Step 2: Create Schema
// The schema is encoded in a string
val schemaString = "name age"
// Generate the schema based on the string of schema
val fields = schemaString.split(" ").map(fieldName => StructField(fieldName, StringType, nullable = true))
// fields.foreach(println)
val schema = StructType(fields)

// Step 3: Apply the schema to the RDD
val peopleDF = spark.createDataFrame(rowRDD, schema)
peopleDF.show()
// Creates a temporary view using the DataFrame
peopleDF.createOrReplaceTempView("people")

// SQL can be run over a temporary view created using DataFrames
val results = spark.sql("SELECT name FROM people WHERE age>20")

// The results of SQL queries are DataFrames and support all the normal RDD operations
// The columns of a row in the result can be accessed by field index or by field name
results.map(attributes => "Name: " + attributes(0)).show()
```
