## 4. RDD Programming Examples

### Ex 1: Word Count

```scala
// Note this is a simple demo without NLP pre-processing steps, such as Stop words removal, Stemming, Tokenization, etc. 
// Please refer to NLP material for a better outcome.

// textFile to load shakespeare.txt
val lines = sc.textFile("shakespeare.txt")
// rddWC1 to store words, but used a filter tranformation to have all the words in lower case.
val rddWC1 = lines.flatMap(line => line.split(" ")).filter(w => w == w.toLowerCase())
// rddWC2 to store paired RDD, the values are all 1s.
val rddWC2 = rddWC1.map(word => (word, 1))
// rddCount to store the final result, it's triggering the action: reduceByKey(func)
// func is to take two inputs and outputs their summation.
val rddCount = rddWC2.reduceByKey((a, b) => (a+b))

// display the sorted result in the descending order of value
// _._2 means the paired RDD's second element - values
// false is for descending order
rddCount.sortBy(_._2,false).collect.foreach(println)

// you can skip all the intermediate RDDs, just have a long chain of transformations
//lines.flatMap(line => line.split(" ")).filter(w => w==w.toLowerCase()).map(word => (word, 1)).reduceByKey((a,b) =>(a+b)).collect.foreach(println)

// you could find the word is split by " ", there are many words having comma or hyphens, 
// in practical session, you could design a better transformation to count the word occurance.
```

### Ex 2: Average Mark Calculation

```scala
// Calculate averaged marks
// (s123456,78)
// (s123456,80)
// (s123456,65)
// (s123456,90)
// (s654321,80)
// (s654321,40)
// (s654321,50)
// (s654321,90)
// (s654321,80)
rddPairScores.collect.foreach(println)
println("-------------")
val rdd1 = rddPairScores.mapValues(x => (x,1))
rdd1.collect.foreach(println)
//rdd1.values.collect.foreach(println)
println("-------------")
// values x and y are pairs that has two elements (marks, 1)
// x._1 (y._1) is the mark while x._2 (y._2) is 1
val rdd2 = rdd1.reduceByKey((x,y) => (x._1+y._1, x._2+y._2))
rdd2.collect.foreach(println)
val rdd3 = rdd2.mapValues(x => x._1.toDouble/x._2)
println("-------------")
rdd3.collect.foreach(println)
```

### Ex 3: Get Top 5 Values

```scala
// Get top value of sale records
// data is stored in csv format: separation is ","

val n = 5
val lines = sc.textFile("sales.txt")
lines.collect.foreach(println)
val rdd1 = lines.map(l => l.split(","))
//rdd1.collect().foreach(println)

// each element is an Array, index starts with 0
val rdd2 = rdd1.map(_(3))
val rdd3 = rdd2.map(_.toDouble)
//rdd3.collect.foreach(println)
val rdd4 = rdd3.sortBy(a => a,false)
// rdd4.collect.foreach(println)
println("-------------")
println(s"Top $n values are:")
rdd4.take(n).foreach(println)
println("-------------")
//Get Min or Max    
rdd3.sortBy(a => a,false).take(1).foreach(println)
rdd3.sortBy(a => a).take(1).foreach(println)


```

```scala
// Movie Rating Example

val ratingSmall = "movie/ratings_small.csv"
val ratingLatest = "movie/ratings_latest.csv"
val moviesSmall = "movie/movies_small.csv"
val moviesLatest = "movie/movies_latest.csv"
// Movie Rating
val ratingLines = sc.textFile(ratingSmall)
//ratingLines.collect.foreach(println)
// sample: 1,1,4.0,964982703
//         1,3,4.0,964981247
val rdd1 = ratingLines.map(line => line.split(","))
// (1,4.0) & (3, 4.0)
val rdd2 = rdd1.map(x => (x(1).toInt, x(2).toDouble))

// (1, (4.0, 3.0, 5.0)) -> (1, (4.0+3.0+5.0)/3) -> (1, 4.0)
val rdd3 = rdd2.groupByKey().map(d => (d._1.toInt,d._2.sum/d._2.size))

// sortBy returns an Array, not RDD, so need parallelize to load into an RDD.
val rdd4 = sc.parallelize(rdd3.sortBy(x => x._2, false).take(100))

//rdd4.foreach(println)
// movies data sample: 
// 1,Toy Story (1995),Adventure|Animation|Children|Comedy|Fantasy
// 2,Jumanji (1995),Adventure|Children|Fantasy
val movieLines = sc.textFile(moviesSmall)
val rdd5 = movieLines.map(line => line.split(",")).map(x => (x(0).toInt, x(1).toString))
// rdd4: (id, average rate) vs rdd5 (id, title) 
// rdd4.join(rdd5) -> (1, (average rate, title)) -> (1, (title, average_rate))
val top100 = rdd4.join(rdd5).map(x => (x._1, x._2._2, x._2._1))
top100.collect().foreach(println)
```

```scala
// A shorter version
val rdd1 = sc.textFile("movie/ratings_small.csv").map(line => line.split(",")).map(x => (x(1).toInt, x(2).toDouble)).groupByKey().map(d => {
    val avg = d._2.sum/d._2.size
    (d._1,avg)
    })
val top100 = sc.textFile("movie/movies_small.csv").map(line => line.split(",")).map(x => (x(0).toInt, x(1))).join(rdd1).map(x => (x._1, x._2._1, x._2._2)).sortBy(x => x._3, false).take(100)
top100.foreach(println)
```
